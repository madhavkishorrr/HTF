```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('Results_60Train_10pred_allerror_MAPE_Computations_HODMD_Figures.csv')
```

```python
df.columns
```

```python
df.head(1)
```

```python
fig, ax = plt.subplots()

fig.set_size_inches(14,4)
fig.set_dpi(600)
sns.barplot(x='Country', y="MAPE",hue="Threshold", data=df)
plt.xlabel('Countries')
plt.ylabel('Mean Average Percentage Error (%)')
plt.gca().legend().set_title('Error Threshold (%)')
plt.savefig('Countries_MAPE_Error_Threshold.eps', format='eps')
plt.savefig('Countries_MAPE_Error_Threshold.png', format='png')
```

```python
fig, ax = plt.subplots()

fig.set_size_inches(16,4)
fig.set_dpi(600)
sns.barplot(x='Country', y="No of Computation",hue="Threshold", data=df)
plt.xlabel('Countries')
plt.ylabel('No of Computations')

plt.gca().legend().set_title('Error Threshold (%)')
#plt.legend(loc='best')
plt.savefig('perf-box.eps', format='eps')
#plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
#plt.legend(loc='upper center')#, bbox_to_anchor=(1.25, 0.5), ncol=1
plt.legend(bbox_to_anchor=(1.002, 1), loc=2, borderaxespad=0.).set_title('Error Threshold (%)')
plt.savefig('Countries_Computations_Threshold.eps', format='eps')
plt.savefig('Countries_Computations_Threshold.png', format='png')
```
